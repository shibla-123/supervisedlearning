### Feature scaling using standard scaler

```python

scaler=StandardScaler()


x_scaled_train=scaler.fit_transform(x_train)
x_scaled_test=scaler.transform(x_test)
```

## Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,classification_report
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
from sklearn.metrics import accuracy_score
```

```python
model.fit(x_scaled_train,y_train)
y_predict = model.predict(x_scaled_test)
accuracy=accuracy_score(y_test,y_predict)

print("accuracy is",accuracy)
print("classification report is",classification_report(y_test,y_predict))
print("confusion matrix is",confusion_matrix(y_test,y_predict))
```

##### This is a supervised learning algorithm.This is used mainly on binary classification problem.In this breast_cancer dataset logistic regression can be used to classify tumors as malignant(1) or benign(0).Here preprocess the dataset and perform necessary scaling.This calculates a weighted sum of input features.This linear combination passed through sigmoid function to produce a probability.This maps the output to a value between 0 and 1.Here the model predicts benign if the probability is greater than 0.5 and malignant otherwise.Here accuracy is >0.5,So the tumor is 'benign'.

##### Decision tree classifier is a supervised  machine learning algorithm used for bothclassification and regression tasks.It is particularly known for its simplicity and interpretability.A decision tree is a flow chart like structure where each internal node represents decision on ana ttribute.each branch represents the outcome of the test,and each leaf node represents class label.The algorithm splits the dataset into subsets.splitting criteria can vary.The algorithm starts at root node .At each node algorithm selects the attribute that best according to the criteria including Gini Impurity,Entropy,Informationgain.The selected feature becomes a decision node,and the data is split accordingly.This process is recursively applied to each branch of the tree until a stopping condition is met.

## SVM

```python
from sklearn.svm import SVC

svc_model=SVC(kernel='rbf',C=3)
svc_model.fit(x_scaled_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,classification_report

svc_pred = svc_model.predict(x_scaled_test)
print("Accuracy score :",accuracy_score(y_test, svc_pred))
print("Precision : ",precision_score(y_test,svc_pred))
print("Recall : ",recall_score(y_test,svc_pred))
print("F1 score : ",f1_score(y_test,svc_pred))
print("Confusion matrix :")
print(confusion_matrix(y_test,svc_pred))
```

```python
from sklearn.model_selection import cross_val_score,GridSearchCV
scores=cross_val_score(svc_model,x_train,y_train,cv=5,scoring='f1_macro')


mean_f1 = np.mean(scores)

print("5 fold cross validation")
print(f'Mean F1 Score: {mean_f1:.2f}')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1,10,100],          
    'gamma': ['scale', 'auto'], 
}


grid_search = GridSearchCV(estimator=svc_model, param_grid=param_grid, scoring='f1_macro', cv=5)

grid_search.fit(x_scaled_train, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

best_f1_score = grid_search.best_score_
print("Best F1 Score:", best_f1_score)

best_model = grid_search.best_estimator_
print(best_model)
```

##### SVM is a supervised machine learning algorithm used for classification,Regression and outlier detection.This is based on decision boundary.It is particularly powerful for classification tasks and works well with the high dimensional data.This works by finding best hyperplane that separates different classes of datapoints in a dataset. 

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_scaled_train,y_train)
```

```python
knn_pred=knn.predict(x_test)
print("accuracy score:",accuracy_score(knn_pred,y_test))
print("classification report:",classification_report(knn_pred,y_test))
```

### Parameter tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid={
           'n_neighbors': [3, 5, 7, 9, 11],  
}

grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, scoring='accuracy', cv=5)

grid_search.fit(x_scaled_train, y_train)

best_params = grid_search.best_params_
print("Best Number of Neighbors:", best_params['n_neighbors'])
best_model = grid_search.best_estimator_
print(best_model)
y_pred = best_model.predict(x_scaled_test)


accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy with Best Model: {accuracy:.2f}')
```

##### knn is a simple,instance based learning algorithm used for classification and regression tasks.This is a lazy learner.KNN does not explicitly learn a model;instead it memorizes the training data and makes decisions based on the stored instances.The value of k determines how  many neighbours are considered when making a prediction.A smaller k captures local patterns but may be noisy,while a larger k smooths out noise but may miss local structures.There is no training phase.for classification,the algorithm finds the k-nearest instances to the query point and assigns the class that is most common amongst neighbors.fro regressioon,the algorithm finds the k-nearest instances and averages their values to make a prediction.
